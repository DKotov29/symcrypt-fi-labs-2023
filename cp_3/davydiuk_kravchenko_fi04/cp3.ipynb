```python
b_lang=["ст", "но", "то", "на", "ен"]
b_text=["тд","рб","во","щю","кд"]

alph1 = "абвгдежзийклмнопрстуфхцчшщыьэюя"
alph2 = "абвгдежзийклмнопрстуфхцчшщьыэюя"
m=31
m2=m*m
ALPH=alph2
```

```python
def extended_euclidian(a, b):
    r_0, r_1 = a, b
    u_0, u_1 = 1, 0
    v_0, v_1 = 0, 1
    while r_1 != 0:
        q = r_0 // r_1
        r_0, r_1 = r_1, r_0 - q * r_1
        u_0, u_1 = u_1, u_0 - q * u_1
        v_0, v_1 = v_1, v_0 - q * v_1
    return r_0, u_0, v_0

def inverse(a, mod):
    gcd, x0, _ = extended_euclidian(a, mod)
    if gcd == 1:
        return x0 % mod
    else:
        print("PANIC")
        return 0

def solve_lin_congruence(a, b, mod):
    a %= mod
    b %= mod
    d, x, _ = extended_euclidian(a, mod)
    if d==1:
        return [(x * b) % mod]
    elif b%d==0:
        a //= d
        b //= d
        mod //= d
        return [((inverse(a, mod) * b) % mod) + k * mod for k in range(d)]
    else:
        return []

def get_bigram_by_number(number):
    return ALPH[number // m] + ALPH[number % m]

def coi(txt):
    k = 0
    lenn = len(txt)
    for t in ALPH:
        ika = 0
        for i in txt:
            if i == t:
                ika += 1
        k += (ika * (ika - 1))
    return float(k / (lenn * (lenn - 1)))

def b_num(bigram):
    return ALPH.index(bigram[0])*m+ALPH.index(bigram[1])
```

```python
def decrypt_bigram(txt, a_inv, b):
    bbb = b_num(txt)
    if bbb is None:
        return ''
    return get_bigram_by_number(a_inv * (bbb - b) % m2)

def decrypt_text(txt, key):
    (a, b) = key
    (d, a_inv, _) = extended_euclidian(a, m2)
    if d != 1:
        return ""
    res = ""
    for i in range(1, len(txt), 2):
        res += decrypt_bigram(txt[i - 1] + txt[i], a_inv, b)
    return res
```

```python
n=5

def find_keys(b_text):
  keys = []
  for a_1 in range(n):
    for b_1 in range(n):
      for a_2 in range(n):
        for b_2 in range(n):
          if a_1 == a_2 or b_1 == b_2:
              continue
          a1 = b_num(b_lang[a_1])
          b1 = b_num(b_text[b_1])
          A = solve_lin_congruence(a1 - b_num(b_lang[a_2]), b1 - b_num(b_text[b_2]), m2)
          keys += [(a, (b1 - a * a1) % m2) for a in A]
  return keys

text = open('/content/text.txt', 'r', encoding="utf8").read().replace('\n', '')

for key in find_keys(b_text):
  decoded = decrypt_text(text, key)
  if decoded == "":
    continue
  norm_I=0.05492369
  if abs(coi(decoded)- norm_I) < 0.01:
    print(decoded)
    print(key)
    break
```
